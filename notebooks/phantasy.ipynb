```python
import pandas as pd
import os
```

```python
dir_path = '../data/phantasy/'

all_entries = os.listdir(dir_path)
filenames = []

for entry in all_entries:
    full_path = os.path.join(dir_path, entry)
    if os.path.isfile(full_path):
        filenames.append(entry)
```

```python
dataframes = []

for f in filenames:
    df = pd.read_csv(dir_path+f)
    dataframes.append(df)
    
final_df = pd.concat(dataframes, axis=0, ignore_index=True)  
```

```python
final_df.to_csv('phantasy.csv')
```
